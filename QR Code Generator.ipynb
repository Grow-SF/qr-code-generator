```python
import qrcode  # https://github.com/lincolnloop/python-qrcode
from qrcode.image.styledpil import StyledPilImage
from qrcode.image.styles.moduledrawers import RoundedModuleDrawer
from qrcode.image.styles.moduledrawers import SquareModuleDrawer
from qrcode.image.styles.colormasks import SolidFillColorMask
```

```python
qr = qrcode.QRCode(
    version=None,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=20,
    border=4,
)
qr.add_data('https://bit.ly/3Kry02m')
qr.make(fit=True)

img = qr.make_image(
    image_factory=StyledPilImage,
    color_mask=SolidFillColorMask(front_color=(0,0,0,255), back_color=(255,255,255,0)),
    embeded_image_path="/Users/sbuss/Downloads/logomark-primary-whitebg-border.png"
)
```

```python
img.save("qr.png")
img
```
