```python
import qrcode  # https://github.com/lincolnloop/python-qrcode
from qrcode.image.styledpil import StyledPilImage
from qrcode.image.styles.moduledrawers import RoundedModuleDrawer
from qrcode.image.styles.moduledrawers import SquareModuleDrawer
from qrcode.image.styles.colormasks import SolidFillColorMask
```

```python
def make(url, outpath):
    qr = qrcode.QRCode(
        version=None,
        error_correction=qrcode.constants.ERROR_CORRECT_H,
        box_size=20,
        border=0,
    )
    qr.add_data(url)
    qr.make(fit=True)

    img = qr.make_image(
        image_factory=StyledPilImage,
        color_mask=SolidFillColorMask(front_color=(0,0,0), back_color=(255,255,255)),
        embeded_image_path="./growsf-logomark-white-bg.png"
    )
    img.save(outpath)
    return img
```

```python
make('https://bit.ly/growsf-nov22-d2', 'd2-nov22.png')
make('https://bit.ly/growsf-d4-nov22', 'd4-nov22.png')
make('https://bit.ly/growsf-d6-nov22', 'd6-nov22.png')
make('https://bit.ly/growsf-d8-nov22', 'd8-nov22.png')
make('https://bit.ly/growsf-d10-nov22', 'd10-nov22.png')
make('https://bit.ly/growsf-citywide-nov22', 'citywide-nov22.png')
make('https://bit.ly/growsf-fullguide-nov22', 'fullguide-nov22.png')
```
